# Visualization with Matplotlib and Seaborn

```python
#Import  library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb 
sb.set_style('darkgrid')
# ensures that plot is shown inline[output in jupyter]
%matplotlib inline  
```

```python
data=pd.read_pickle('data')
```

# To check missing values

```python
sb.heatmap(data.isnull(),cbar=False,yticklabels=False,cmap='viridis')
```

* Dataset has no missing values.
* If there were any, you would've noticed in figure represented by different colour shade on purple background.
* Do try it out with other dataset which has missing values,you'll see the difference.

```python
plt.figure(figsize=(17,4))
#sns.heatmap(df.corr(),cmap='Blues',annot=False) 
data.notnull().sum().plot('bar')
```

## Target data

```python
data.PerformanceRating.value_counts().plot('bar')
plt.ylabel('Numner of employee Rating Wise',fontsize=15,color='blue')
plt.xlabel('Rating',fontsize=15,color='blue')
plt.title("Perforance Rating",fontsize=20)
```

## Department wise performances

```python
plt.figure(figsize=(10,4))
x=dep_per=data.groupby(data['EmpDepartment']).mean().loc[:,['PerformanceRating']]
plt.plot(x,marker='*',markersize=10)
```

## To check correlation

```python
plt.figure(figsize=(10,10))
sb.heatmap(data.corr(),cmap='Blues',annot=False)
```

## Analyzing

```python
#check outlier
plt.figure(figsize=(14,3))
sb.catplot(x="EmpDepartment", y="Age", kind="box", data=data)
```

```python
plt.figure(figsize=(9,4))
sb.countplot(x="EmpDepartment", data=data)
plt.show()
```

```python
plt.figure(figsize=(9,4))
sb.countplot(x="EmpDepartment",data=data)
plt.show()
```

```python
plt.figure(figsize=(9,4))
sb.countplot(x="EmpDepartment", hue='Gender',data=data)
plt.show()
```

## observation :
* In Sales department more employees working as compared to all department.
* Number of male employee is greatter than female employee.

```python

plt.figure(figsize=(10,4))
plt.title('EducationBackground v/s PerformanceRating')
sb.countplot(x='EducationBackground',hue='PerformanceRating',data=data)
```

```python
sb.barplot(x='EducationBackground',hue="OverTime",y='PerformanceRating',data=data)
plt.title('Education Background Vs PerformanceRating',fontsize=19)
plt.legend(bbox_to_anchor=(1.21,1),loc=2) #Put the legend Out side the figure
```

### observation:
If employees works as overtime Rating will be improve.

```python
plt.figure(figsize=(16,4))
plt.title('PerformanceRating Age wise',fontsize=19,color='red')
sb.barplot(x='Age',y='PerformanceRating',data=data)
```

```python
plt.figure(figsize=(16,4))
plt.title('PerformanceRating v\s EmpHourlyRate',fontsize=19,color='red')
sb.barplot(x='EmpHourlyRate',y='PerformanceRating',data=data)
```
